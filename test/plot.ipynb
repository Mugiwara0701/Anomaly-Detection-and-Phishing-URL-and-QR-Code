```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

plt.style.use('seaborn-v0_8-white')
```

```python
plt.style.available
```

```python
def plot_phishing_trend():
    quarters = ['Q3 2020', 'Q4 2020', 'Q1 2021', 'Q3 2021', 'Q1 2022', 'Q2 2022', 'Q1 2023', 'Q3 2023', 'Q1 2024']
    phishing_sites = [165772, 571764, 611877, 730372, 1025968, 1270883, 999956, 963994, 1624144]

    plt.figure(figsize=(10, 6))
    plt.bar(quarters, phishing_sites, color='teal', alpha=0.7)
    plt.title('Trend of Unique Phishing Sites Over Time (Bar Graph)')
    plt.xlabel('Quarter')
    plt.ylabel('No. of Unique Phishing Sites')
    plt.xticks(rotation=45)
    for i, v in enumerate(phishing_sites):
        plt.text(i, v, str(v), ha='center', va='bottom')
    plt.tight_layout()
    plt.show()
```

```python
def plot_domain_length_distribution():
    np.random.seed(42)
    domain_lengths_0 = np.random.normal(15, 5, 600000)  # Label 0
    domain_lengths_1 = np.random.normal(12, 4, 400000)  # Label 1

    plt.figure(figsize=(10, 6))
    plt.hist(domain_lengths_0, bins=60, alpha=0.7, label='0', color='blue')
    plt.hist(domain_lengths_1, bins=60, alpha=0.7, label='1', color='grey')
    plt.title('Distribution of domain_length')
    plt.xlabel('domain_length')
    plt.ylabel('Count')
    plt.legend()
    plt.tight_layout()
    plt.show()
```

```python
def plot_validation_accuracy():
    models = ['LSTM', 'CNN', 'GRU', 'BILSTM', 'CNN-GRU', 'ATTENTION', 'RANDOM FOREST']
    accuracies = [99.91, 99.93, 99.93, 99.85, 99.95, 99.95, 99.97]

    plt.figure(figsize=(10, 6))
    plt.bar(models, accuracies, color='lightgreen', alpha=0.7)
    plt.title('Validation Accuracy')
    plt.xlabel('Models')
    plt.ylabel('Accuracy')
    for i, v in enumerate(accuracies):
        plt.text(i, v, f'{v}%', ha='center', va='bottom')
    plt.ylim(99.80, 100)
    plt.tight_layout()
    plt.show()
```

```python
def plot_validation_loss():
    models = ['LSTM', 'CNN', 'GRU', 'BILSTM', 'CNN-GRU', 'ATTENTION', 'RANDOM FOREST']
    losses = [0.012, 0.009, 0.007, 0.011, 0.005, 0.005, 0.003]

    plt.figure(figsize=(10, 6))
    plt.bar(models, losses, color='salmon', alpha=0.7)
    plt.title('Validation Loss')
    plt.xlabel('Models')
    plt.ylabel('Loss')
    for i, v in enumerate(losses):
        plt.text(i, v, f'{v:.4f}', ha='center', va='bottom')
    plt.tight_layout()
    plt.show()
```

```python
def plot_test_accuracy():
    models = ['LSTM', 'CNN', 'GRU', 'BILSTM', 'CNN-GRU', 'ATTENTION', 'RANDOM FOREST']
    accuracies = [99.90, 99.92, 99.93, 99.89, 99.95, 99.95, 99.97]

    plt.figure(figsize=(10, 6))
    plt.bar(models, accuracies, color='lightblue', alpha=0.7)
    plt.title('Test Accuracy')
    plt.xlabel('Models')
    plt.ylabel('Accuracy')
    for i, v in enumerate(accuracies):
        plt.text(i, v, f'{v}%', ha='center', va='bottom')
    plt.ylim(99.80, 100)
    plt.tight_layout()
    plt.show()
```

```python
def plot_test_loss():
    models = ['LSTM', 'CNN', 'GRU', 'BILSTM', 'CNN-GRU', 'ATTENTION', 'RANDOM FOREST']
    losses = [0.031, 0.020, 0.020, 0.028, 0.016, 0.015, 0.010]

    plt.figure(figsize=(10, 6))
    plt.plot(models, losses, marker='o', color='purple', label='Test Loss')
    plt.title('Test Loss Across Different Models')
    plt.xlabel('Models')
    plt.ylabel('Test Loss')
    for i, v in enumerate(losses):
        plt.text(i, v, f'{v:.4f}', ha='center', va='bottom')
    plt.legend()
    plt.show()
```

```python
def plot_false_positive_rate():
    models = ['LSTM MODEL', 'CNN MODEL', 'GRU MODEL', 'BILSTM MODEL', 'CNN-GRU MODEL', 'ATTENTION MODEL', 'RANDOM FOREST']
    rates = [0.012, 0.009, 0.007, 0.011, 0.005, 0.005, 0.001]

    plt.figure(figsize=(10, 6))
    plt.bar(models, rates, color='coral', alpha=0.7)
    plt.title('False Positive Rate Comparison')
    plt.xlabel('Models')
    plt.ylabel('False Positive Rate')
    for i, v in enumerate(rates):
        plt.text(i, v, f'{v:.4f}', ha='center', va='bottom')
    plt.tight_layout()
    plt.show()
```

```python
plot_phishing_trend()
plot_domain_length_distribution()
plot_validation_accuracy()
plot_validation_loss()
plot_test_accuracy()
plot_test_loss()
plot_false_positive_rate()
```
