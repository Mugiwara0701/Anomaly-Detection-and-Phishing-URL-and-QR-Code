```python
import pandas as pd
import numpy as np
from sklearn.ensemble import IsolationForest
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

df = pd.read_csv('../backend/data/network_data.csv')

df = df.drop('timestamp', axis=1)

contamination = 0.1 

X = df[['download', 'upload', 'latency', 'packets_received', 'cpu', 'memory']]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_train, X_test = train_test_split(X_scaled, test_size=0.2, random_state=42)

iso_forest = IsolationForest(contamination=contamination, random_state=42)
iso_forest.fit(X_train)

y_pred_train = iso_forest.predict(X_train)
y_pred_test = iso_forest.predict(X_test)

y_test_labels = np.where(y_pred_test == -1, 1, 0)

y_test_pred = iso_forest.predict(X_test)
y_test_pred_labels = np.where(y_test_pred == -1, 1, 0)

accuracy = accuracy_score(y_test_labels, y_test_pred_labels)
print(f"Accuracy on test set: {accuracy:.2f}")

anomaly_scores = iso_forest.score_samples(X_test)
```
